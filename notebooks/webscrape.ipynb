# Webscrape 

This notebook is used to scrape data from reelgood.com, which includes tv shows and movies available in different streaming services.

Note that through trial and error, it was determined that this was the last page of that list (after filtering for Netflix, Hulu, Amazon Prime, Disney+, HBO): https://reelgood.com/all?offset=24592

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from pprint import pprint
```

stackoverflow link that may be helpful:
https://stackoverflow.com/questions/50882622/web-scrapping-multiple-pages-when-the-url-changes-and-adds-offset-here

```python
#using just netflix as a start.
url = 'https://reelgood.com/source/netflix'
```

```python
#save request 
response = requests.get(url)
```

```python
#check status of response.  200 is good
response.status_code
```

```python
#check if something came through as expected. Seems to work
response.content
```

```python
soup = BeautifulSoup(response.content, 'lxml')
```

```python
pprint(soup)
```

```python
table = soup.find_all('table',attrs={'class':'css-1179hly'})
```

```python
pprint(table)
```

```python
#checking that only one table came through, as expected
len(table)
```

```python
result_list = pd.read_html(str(table[0])) # a list of dataframes
```

```python
df = result_list[0]
df.head()
```
